```python
#Divide The Dataset Into 2 parts i.e. Training(70%) and Testing(30%)
from sklearn.model_selection import train_test_split
#Apply It:-
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
X_train.shape,Y_train.shape
```

```python
X_test.shape,Y_test.shape
```

```python
#Now Before Giving The data To the Machine We have to Perform Scaling i.e Standard Scaling on the Dataset:-
from sklearn.preprocessing import StandardScaler
#Create An Object Of The Standard Scaler Class:-
ss=StandardScaler()
#Apply It:-
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#Now Create A User Function To Train And Test The Data in with Every Algorithm Possible:-
def create_model(model):
    #train Model
    model.fit(X_train,Y_train)
    #test Model
    Y_predict=model.predict(X_test)
    #generating The Report:-
    from sklearn.metrics import classification_report
    print(classification_report(Y_test,Y_predict))
    #Generate The Confusion Matrix:-
    from sklearn.metrics import confusion_matrix
    print(confusion_matrix(Y_test,Y_predict))
    return model
```

# SVM:Support Vector Machine
There are different types Of SVM:-
1. Linear Separable data means Linear Kernal function
2. Non -Linear data

# Linear Separable data means Linear Kernal function

```python
from sklearn.svm import LinearSVC #LinearSVC means Linear Support Vector Classifier.
```

```python
#Create an object of LinearSVC:-
svc=LinearSVC(random_state=1) #bydefault hard margin,suppose no outlier in my dataset means suppose no overfit.
#svc is user defined object
```

```python
#Call the function:-
svc=create_model(svc)
```

```python
#here recall is 0.60 of 1 category which is not good but we can do more better
#means can be possible ,model is overfit  why 
#perhaps , can be no error on training time but error on testing time , 
#what  do we do ,
#add some external error during training time  if create a object of 
#LinearSVC class 

#Soft margin means to reduced overfitting situation means some error add on
#training time 
#create object of LinearSVC class
#Soft Margin
```

```python
#Create object of LinearSVC class and add some Externl Error:-
svc1=LinearSVC(random_state=1,C=0.9) #Soft Margin
#here C is parameter which is used to add some error during training time 
#set range of error 0.001 to 1 means C<=1
```

```python
#Call the Function:-
svc1=create_model(svc1)
```

```python
#This Shows That Our Data is Not Linear Separatable Data.
#Because No Matter How Much amount of error we increase there is no change in the model.
```

```python
#no any effect on recall if error increase or decrease means we conclude ,
#data set is not a linearly separateable means non-linearly data 
#Then use polynomial Kernel function : It is used for Non-Linearly data
#In polynomial Kernal function, increase the dimension means to increase low
#dimension to high dimension
```

```python
#Whenever Such Situation Occurs where the data is not linear Separatable then we will give the data to the the kernel function.
#Kernel Function:-There are two types in Kernel Function
#1.Polynomial Kernel Function:-In this situation we will increse the dimensions of the data,
#In This boundary which separates the two classes is called as hyperplane or decision boundary.
#   i.e. if the data is 1 dimensional then we will make it 2 dimensional
#   i.e. if the data is 2 dimensional then we will make it 3 dimensional
#   i.e. if the data is 3 dimensional then we will make it 4 dimensional
#2.Radial Basis Kernel Function:-In this situation if the data is forming a circular pattern.
#In This boundary which separates the two classes is called as hypercircle or decision boundary Circle.
```

```python
#Use polynomial kernel function , use inbuilt class SVC(Support Vector Classifier)
from sklearn.svm import SVC
#SVC used for polynomial and radial basis kerne function: both are used for non-linear data.
```

```python
#Now create an object of the SVC class and passing the poly in kernel parameter:-
svc1=SVC(random_state=1,kernel='poly')
```

```python
#Call Function:-
svc1=create_model(svc1)
```

```python
#Score=43% its not good
#So give the dataset in radial basis kernel function:-
#Create object of SVC class:-
svc2=SVC(random_state=1,kernel='rbf') #rbf means radius basis function
```

```python
#Call Function:-
svc2=create_model(svc2)
```
